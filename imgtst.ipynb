```python
from bing_image_downloader import downloader
```

```python
query_string=[#'F-15 Eagle',
              #'F-22 Raptor',
              #'F-35 Lightning II',
              #'F-16 Fighting Falcon',
              #'A-10 Thunderbolt II','AC-130 Ghostrider','B-1 Lancer','B-2 Spirit',
              #'B-52 Stratofortress','E-3 Sentry',
    #'E-4 Nightwatch','E-11A','EA-37B',
              'EC-130 Commando Solo',
    'EC-130H Compass Call','U-2 Dragon Lady',
              'RC-135 Rivet Joint','KC-135 Stratotanker','C-17 Globemaster III',
              'C-130J Super Hercules','V-22 Osprey','RQ-4 Global Hawk',
              'MQ-9 Reaper']
```

```python
def query(a):
    for x in a:
        downloader.download(x.lower(), limit=100,  output_dir='data/img_clss/', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
    
```

```python
query(query_string)
```
